```python
import pandas as pd

df = pd.read_excel("https://docs.google.com/spreadsheets/d/1nxcPuu1PYQEs3RYTiUKlLldaMkBb4FdqdtGa8GeOwx4/export?format=xlsx")
print(df)
```

```python
dimensao_chapa = 5.04

areas = []

for index, linha in df.iterrows():
    area = float(linha["Largura (m)"] * linha["Altura (m)"])
    areas.append(area)
    print(area)

print(areas)
```

```python
for i in range(len(areas)):
    quantidade_chapas = (dimensao_chapa / areas[i])
    sobras = quantidade_chapas - int(quantidade_chapas)
    print(quantidade_chapas)
    print(sobras)

```
