```python
import pandas as pd

print("Baixando os dados da planilha...")
# 1. Lendo os dados da planilha
url = "https://docs.google.com/spreadsheets/d/17yC9PcTkC3gY5orbQ5VQXeBDOStMx2Dfpo53cMZTn5k/export?format=xlsx"
df_tipos_de_vigas = pd.read_excel(url, sheet_name="Tipos de vigas")
df_capacidade_das_formas = pd.read_excel(url, sheet_name="Capacidade das fôrmas")

# 2. Pegando o tamanho da fôrma e criando as peças "no chão"
tamanho_forma = df_capacidade_das_formas['Capacidade (m)'].max()

pecas_no_chao = []
for indice, linha in df_tipos_de_vigas.iterrows():
    tamanho = linha['Tamanho (m)']
    demanda = int(linha['Demanda'])
    for vez in range(demanda):
        pecas_no_chao.append(tamanho)

print(f"Tamanho da fôrma base: {tamanho_forma}m")
print(f"Total de peças para organizar: {len(pecas_no_chao)} vigas.")
print("-" * 40)

# ========================================================
# INÍCIO DO MÉTODO GULOSO (FIRST-FIT DECREASING)
# ========================================================

# Passo 1: A Fila Indiana (Ordena do MAIOR para o MENOR sozinho)
pecas_no_chao.sort(reverse=True)

# Passo 2: A lista onde vamos guardar as fôrmas prontas
formas_usadas = []

# Passo 3: Vamos pegar uma peça de cada vez da fila
for peca in pecas_no_chao:
    
    guardou = False # Começa avisando que ainda não guardamos essa peça
    
    # Vamos olhar para todas as fôrmas que já abrimos e estão em uso
    for forma in formas_usadas:
        espaco_ocupado = sum(forma) # Soma o tamanho das peças que já estão lá dentro
        
        # Se colocar essa peça nova não estourar o limite de 11.95m...
        if espaco_ocupado + peca <= tamanho_forma:
            forma.append(peca) # Guarda a peça na fôrma
            guardou = True     # Avisa que deu certo!
            break              # Para de procurar outra fôrma
            
    # Se a gente olhou TODAS as fôrmas abertas e não coube em nenhuma...
    if guardou == False:
        # Abre uma fôrma NOVA já com essa peça lá dentro
        formas_usadas.append([peca])

# ========================================================
# FIM DO MÉTODO - IMPRIMINDO OS RESULTADOS
# ========================================================

desperdicio_total = 0

print(f"\n--- RESULTADO FINAL DO PLANO DE CORTE ---")
print(f"Total de fôrmas utilizadas: {len(formas_usadas)}\n")

# Para mostrar como cada fôrma ficou montada
for i, forma in enumerate(formas_usadas):
    uso = sum(forma)
    sobra = tamanho_forma - uso
    desperdicio_total = desperdicio_total + sobra
    
    # Imprime a fôrma com os números arredondados para ficar bonito
    print(f"Fôrma {i+1:02d}: {forma} | Uso: {uso:.2f}m | Sobra: {sobra:.2f}m")

print("-" * 40)
print(f"Desperdício total (soma de todas as sobras): {desperdicio_total:.2f}m")
```
