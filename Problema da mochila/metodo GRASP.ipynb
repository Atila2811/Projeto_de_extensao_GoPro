```python
import pandas as pd
from tabulate import tabulate

url = "https://docs.google.com/spreadsheets/d/1jUUx-ZZlHVs8n8byGHPm7L5-kHLsqlFW/export?format=xlsx"

df_mochila = pd.read_excel(url)
print(df_mochila)
```

# Problema da Mochila com Método GRASP

Este notebook implementa uma solução para o **Problema da Mochila Multidimensional** (Multi-Constrained Knapsack Problem) utilizando a metaheurística **GRASP** (Greedy Randomized Adaptive Search Procedure).

## O Problema da Mochila
O Problema da Mochila é um clássico problema de otimização combinatória onde, dado um conjunto de itens (projetos), cada um com um valor (retorno) e múltiplos pesos (capital e mão de obra), deve-se maximizar o valor total selecionando itens respeitando as restrições de capacidade.

## Método GRASP
O GRASP é uma metaheurística iterativa composta por duas fases principais:

1. **Fase de Construção Greedy Randomizada**: Constrói uma solução inicial selecionando elementos de forma gulosa, mas com aleatoriedade controlada por um parâmetro $\alpha$.
2. **Fase de Busca Local** (Local Search): Melhora a solução construída através de pequenas perturbações (troca, inserção, remoção de itens).

## Dados do Problema
- **Projetos**: Itens candidatos a serem selecionados
- **Retorno ($v_i$)**: Valor/benefício de cada projeto
- **Capital ($w_i$)**: Custo financeiro necessário
- **Mão de obra ($a_i$)**: Recurso humano necessário

## Restrições
- Capital total $\leq 500$
- Mão de obra total $\leq 600$

## Objetivo
Maximizar: $\sum_{i \in S} v_i$

Sujeito a:
- $\sum_{i \in S} w_i \leq 500$
- $\sum_{i \in S} a_i \leq 600$

Onde $S$ é o conjunto de projetos selecionados.

```python
eficiencias = []

for index, linha in df_mochila.iterrows():
    eficiencia = float(linha["Retorno (vᵢ)"] / (linha["Capital (wᵢ)"] + linha["Mão de obra (aᵢ)"]))
    eficiencias.append(eficiencia)

print(eficiencias)

df_eficiencia = pd.DataFrame({"Eficiencia": eficiencias})
print(f"DataFrame de Eficiências:")
print(tabulate(df_eficiencia, headers="keys", tablefmt="fancy_grid"))
```

## 2. Cálculo da Eficiência (Critério Guloso)

Para a fase de construção do GRASP, precisamos de uma métrica para avaliar a qualidade de cada projeto. Utilizamos a **Eficiência** como critério guloso:

$$
\text{Eficiência}_i = \frac{v_i}{w_i + a_i}
$$

Onde:
- $v_i$: Retorno (valor) do projeto $i$
- $w_i$: Capital necessário
- $a_i$: Mão de obra necessária

Esta razão representa o retorno por unidade de recurso total consumido. Projetos com maior eficiência são prioritários na fase de construção da solução.

```python
df_mochila = pd.concat([df_mochila, df_eficiencia], axis=1)
print(f"DataFrame Completo com Eficiências:")
print(tabulate(df_mochila, headers="keys", tablefmt="fancy_grid"))

df_mochila_sorted = df_mochila.sort_values(by="Eficiencia", ascending=False)
df_mochila_sorted.index += 1
print(df_mochila_sorted)
```

## 3. Ordenação dos Projetos por Eficiência

Anexamos a coluna de eficiências ao DataFrame e ordenamos os projetos de forma **decrescente** pela eficiência.

Esta ordenação define a **lista de candidatos** para a fase de construção do GRASP. Projetos mais eficientes aparecem primeiro, servindo como base para a seleção gulosa.

A ordenação segue o critério:

$$
\text{Projetos ordenados por } \text{Eficiência}_i \text{ (maior para menor)}
$$

Isso garante que a construção da solução comece pelos projetos com melhor custo-benefício.

```python
capital_total = 0
mao_de_obra_total = 0

projetos_selecionados = []
Retorno_total = 0

for index, linha in df_mochila_sorted.iterrows():
    novo_capital = capital_total + linha["Capital (wᵢ)"]
    nova_mao_de_obra = mao_de_obra_total + linha["Mão de obra (aᵢ)"]
    novo_retorno = Retorno_total + linha["Retorno (vᵢ)"]

    if novo_capital <= 500 and nova_mao_de_obra <= 600:
        capital_total = novo_capital
        mao_de_obra_total = nova_mao_de_obra
        Retorno_total = novo_retorno
        projeto = int(linha["Projeto"])
        projetos_selecionados.append(projeto)

print(f"\nProjetos selecionados: {sorted(projetos_selecionados)}")
print(f"Capital total utilizado: {capital_total}")
print(f"Mão de obra total utilizada: {mao_de_obra_total}")
print(f"Total de projetos selecionados: {len(projetos_selecionados)}")
print(f"Retorno total: {Retorno_total}")
```

## 4. Fase de Construção Gulosa (GRASP - Fase 1)

Nesta etapa, percorremos os projetos ordenados por eficiência e selecionamos aqueles que ainda respeitam as restrições de capacidade.

### Critério de Aceitação
Um projeto $i$ é selecionado se:

$$
\text{capital\_total} + w_i \leq 500 \quad \text{e} \quad \text{mão\_de\_obra\_total} + a_i \leq 600
$$

### Atualização dos Recursos
Quando um projeto é aceito:
- $\text{capital\_total} \leftarrow \text{capital\_total} + w_i$
- $\text{mão\_de\_obra\_total} \leftarrow \text{mão\_de\_obra\_total} + a_i$
- $\text{retorno\_total} \leftarrow \text{retorno\_total} + v_i$

### Resultado
Ao final, obtemos:
- **Conjunto $S$**: Projetos selecionados
- **Retorno total**: $\sum_{i \in S} v_i$
- **Recursos consumidos**: Capital e mão de obra utilizados

**Nota**: Esta é uma implementação da fase gulosa do GRASP. Uma versão completa incluiria também a parte de Busca Local (troca de projetos, inserção/remoção aleatória) para melhorar a solução.
