# Problema da Mochila com MГ©todo GRASP

Este notebook implementa uma soluГ§ГЈo para o **Problema da Mochila Multidimensional** (Multi-Constrained Knapsack Problem) utilizando a metaheurГӯstica **GRASP** (Greedy Randomized Adaptive Search Procedure).

## O Problema da Mochila
O Problema da Mochila Г© um clГЎssico problema de otimizaГ§ГЈo combinatГіria onde, dado um conjunto de itens (projetos), cada um com um valor (retorno) e mГәltiplos pesos (capital e mГЈo de obra), deve-se maximizar o valor total selecionando itens respeitando as restriГ§Гөes de capacidade.

## MГ©todo GRASP
O GRASP Г© uma metaheurГӯstica iterativa composta por duas fases principais:

1. **Fase de ConstruГ§ГЈo Greedy Randomizada**: ConstrГіi uma soluГ§ГЈo inicial selecionando elementos de forma gulosa, mas com aleatoriedade controlada por um parГўmetro $\alpha$.
2. **Fase de Busca Local** (Local Search): Melhora a soluГ§ГЈo construГӯda atravГ©s de pequenas perturbaГ§Гөes (troca, inserГ§ГЈo, remoГ§ГЈo de itens).

## Dados do Problema
- **Projetos**: Itens candidatos a serem selecionados
- **Retorno ($v_i$)**: Valor/benefГӯcio de cada projeto
- **Capital ($w_i$)**: Custo financeiro necessГЎrio
- **MГЈo de obra ($a_i$)**: Recurso humano necessГЎrio

## RestriГ§Гөes
- Capital total $\leq 500$
- MГЈo de obra total $\leq 600$

## Objetivo
Maximizar: $\sum_{i \in S} v_i$

Sujeito a:
- $\sum_{i \in S} w_i \leq 500$
- $\sum_{i \in S} a_i \leq 600$

Onde $S$ Г© o conjunto de projetos selecionados.

```python
eficiencias = []

for index, linha in df_mochila.iterrows():
    eficiencia = float(linha["Retorno (vбөў)"] / (linha["Capital (wбөў)"] + linha["MГЈo de obra (aбөў)"]))
    eficiencias.append(eficiencia)

print(eficiencias)

df_eficiencia = pd.DataFrame({"Eficiencia": eficiencias})
print(f"DataFrame de EficiГӘncias:")
print(tabulate(df_eficiencia, headers="keys", tablefmt="fancy_grid"))
```

## 2. CГЎlculo da EficiГӘncia (CritГ©rio Guloso)

Para a fase de construГ§ГЈo do GRASP, precisamos de uma mГ©trica para avaliar a qualidade de cada projeto. Utilizamos a **EficiГӘncia** como critГ©rio guloso:

$$
\text{EficiГӘncia}_i = \frac{v_i}{w_i + a_i}
$$

Onde:
- $v_i$: Retorno (valor) do projeto $i$
- $w_i$: Capital necessГЎrio
- $a_i$: MГЈo de obra necessГЎria

Esta razГЈo representa o retorno por unidade de recurso total consumido. Projetos com maior eficiГӘncia sГЈo prioritГЎrios na fase de construГ§ГЈo da soluГ§ГЈo.

```python
df_mochila = pd.concat([df_mochila, df_eficiencia], axis=1)
print(f"DataFrame Completo com EficiГӘncias:")
print(tabulate(df_mochila, headers="keys", tablefmt="fancy_grid"))

df_mochila_sorted = df_mochila.sort_values(by="Eficiencia", ascending=False)
df_mochila_sorted.index += 1
print(df_mochila_sorted)
```

## 3. OrdenaГ§ГЈo dos Projetos por EficiГӘncia

Anexamos a coluna de eficiГӘncias ao DataFrame e ordenamos os projetos de forma **decrescente** pela eficiГӘncia.

Esta ordenaГ§ГЈo define a **lista de candidatos** para a fase de construГ§ГЈo do GRASP. Projetos mais eficientes aparecem primeiro, servindo como base para a seleГ§ГЈo gulosa.

A ordenaГ§ГЈo segue o critГ©rio:

$$
\text{Projetos ordenados por } \text{EficiГӘncia}_i \text{ (maior para menor)}
$$

Isso garante que a construГ§ГЈo da soluГ§ГЈo comece pelos projetos com melhor custo-benefГӯcio.

```python
capital_total = 0
mao_de_obra_total = 0

projetos_selecionados = []
Retorno_total = 0

for index, linha in df_mochila_sorted.iterrows():
    novo_capital = capital_total + linha["Capital (wбөў)"]
    nova_mao_de_obra = mao_de_obra_total + linha["MГЈo de obra (aбөў)"]
    novo_retorno = Retorno_total + linha["Retorno (vбөў)"]

    if novo_capital <= 500 and nova_mao_de_obra <= 600:
        capital_total = novo_capital
        mao_de_obra_total = nova_mao_de_obra
        Retorno_total = novo_retorno
        projeto = int(linha["Projeto"])
        projetos_selecionados.append(projeto)

print(f"\nProjetos selecionados: {sorted(projetos_selecionados)}")
print(f"Capital total utilizado: {capital_total}")
print(f"MГЈo de obra total utilizada: {mao_de_obra_total}")
print(f"Total de projetos selecionados: {len(projetos_selecionados)}")
print(f"Retorno total: {Retorno_total}")
```

## 4. Fase de ConstruГ§ГЈo Gulosa (GRASP - Fase 1)

Nesta etapa, percorremos os projetos ordenados por eficiГӘncia e selecionamos aqueles que ainda respeitam as restriГ§Гөes de capacidade.

### CritГ©rio de AceitaГ§ГЈo
Um projeto $i$ Г© selecionado se:

$$
\text{capital\_total} + w_i \leq 500 \quad \text{e} \quad \text{mГЈo\_de\_obra\_total} + a_i \leq 600
$$

### AtualizaГ§ГЈo dos Recursos
Quando um projeto Г© aceito:
- $\text{capital\_total} \leftarrow \text{capital\_total} + w_i$
- $\text{mГЈo\_de\_obra\_total} \leftarrow \text{mГЈo\_de\_obra\_total} + a_i$
- $\text{retorno\_total} \leftarrow \text{retorno\_total} + v_i$

### Resultado
Ao final, obtemos:
- **Conjunto $S$**: Projetos selecionados
- **Retorno total**: $\sum_{i \in S} v_i$
- **Recursos consumidos**: Capital e mГЈo de obra utilizados

**Nota**: Esta Г© uma implementaГ§ГЈo da fase gulosa do GRASP. Uma versГЈo completa incluiria tambГ©m a parte de Busca Local (troca de projetos, inserГ§ГЈo/remoГ§ГЈo aleatГіria) para melhorar a soluГ§ГЈo.
